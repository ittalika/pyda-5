```python
import pandas as pd

data_rating = pd.read_csv('ratings.csv')
rating = data_rating['rating']

def classify_by(rating):
    
    if rating <= 2.0:
        rating_class = 'низкий рейтинг'
    elif 2 < rating <= 4.0:
        rating_class = 'средний рейтинг'
    else:
        rating_class = 'высокий рейтинг'
    
    return rating_class
    
data_rating['class'] = rating.apply(classify_by)

data_rating.tail(20)
```

```python
import re

data_keywords = pd.read_csv('keywords.csv')
keyword = data_keywords['keyword']

geo_data = {

'Центр': [r'москв[а-я]+', r'тул[а-я]+', r'ярославл[а-я]+'],
'Северо-Запад': [r'петербург[а-я]+', r'псков[а-я]+', r'мурманск[а-я]+'],
'Дальний Восток': [r'владивосток[а-я]+', r'сахалин[а-я]+', r'хабаровск[а-я]+']}

def geo_class(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if re.findall(city, keyword):
                return region
    else: 
        return 'undefined'
            
data_keywords['region'] = keyword.apply(geo_class)

data_keywords[ ['region', 'keyword'] ].groupby('region').count().reset_index().sort_values(by='keyword', ascending=False)
```

```python
data_rating = pd.read_csv('ratings.csv')
data_movies = pd.read_csv('movies.csv')

title = data_movies['title']

years = [str(x) for x in range(1950, 2011)]

def production_year(title):
    year = ''.join(re.findall(r'\d{4}', title))
    if year in years:
        return year
    else:
        return '1900'

data_movies['year'] = title.apply(production_year)
    
data_joined = data_rating.merge(data_movies, on='movieId', how='left')

data_joined.groupby('year').mean()['rating'].sort_values(ascending=False).head(10)
```
