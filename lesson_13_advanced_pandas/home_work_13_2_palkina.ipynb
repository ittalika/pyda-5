```python
import pandas as pd

ratings = pd.read_csv('ratings.csv')

res = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'userId': 'count'})
res_above_100 = res.loc[res['userId']['count'] >= 100]

diff = res_above_100['timestamp']['max'] - res_above_100['timestamp']['min']
diff.mean()

print(round(diff.mean()/2.628e+6), 'месяцев')
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)

client_base_no_adrs = client_base.loc[:, ['client_id']]

table_1 = client_base_no_adrs.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')
table_2 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')
```

В зависимости от задачи можно работать с геоданными как со списками или словарями (в последнем случае ключом может быть широта, а значением долгота). Для сквозной аналитики можно объединить таблицы по user_id, предварительно обдумав вопрос с дубликатами. Если у пользователя несколько координат, то опять же в зависимости от задачи для последующих расчетов можно использовать среднее, первое, последнее или самое частое значение. Использовать координаты можно в рамках расширений pandas, например, библиотеки geopy, которая поможет определить не только город и страну, но и точный адрес. Некоторые другие инструменты также позволяют визуализировать точки на карте.
