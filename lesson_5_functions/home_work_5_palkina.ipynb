```python
documents = [
{"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
{"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
{"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
"1": ["2207 876234", "11-2"],
"2": ["10006"],
"3": []
}

def doc_owner():
    print()  
    print("Введите номер документа: ")
    user_input = input()
    print()
    for doc in documents:
        if user_input == doc["number"]:
            owner = doc["name"]
            break
    else: 
        owner = "Документ в базе не найден"
    print("Результат:", owner)
doc_owner()

def doc_shelf():
    print()  
    print("Введите номер документа: ")
    user_input = input()
    print()
    num_list = list(directories.values())
    i = 0
    for dir in directories:  
        if user_input in num_list[i]:
            shelf = i + 1
            print("Результат:")
            print("Документ хранится на полке:", shelf)
            break
        else:
            i += 1
    else:
        print("Результат:", "Документ в базе не найден", sep="\n")
doc_shelf()

def all_doc():
    print()
    print("Результат:")
    num_list = list(directories.values())
    for doc in documents:
        for i in range(0, 3):
            if doc["number"] in num_list[i]:
                shelf = i + 1
                print("№: {}, тип: {}, владелец: {}, полка хранения: {}".format(doc['number'], doc['type'], doc['name'], shelf))
all_doc()

def new_shelf():
    print()  
    print("Введите номер полки: ")
    user_input = input()
    print()
    if user_input in list(directories):
        print("Результат:")
        print("Такая полка уже существует. Текущий перечень полок: ", ", ".join(list(directories))) 
    else:
        directories[user_input] = []
        print("Результат:")
        print("Полка добавлена. Текущий перечень полок: ", ", ".join(list(directories)))
new_shelf()

def delete_shelf():
    print()  
    print("Введите номер полки: ")
    user_input = input()
    print()
    if user_input in list(directories) and directories[user_input] == []:
        del(directories[user_input])
        print("Результат: ") 
        print("Полка удалена. Текущий перечень полок:", ", ".join(list(directories)))
    elif user_input in list(directories) and directories[user_input] != []:
        print("Результат: ") 
        print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:", ", ".join(list(directories)))
    else:
        print("Результат: ") 
        print("Такой полки не существует. Текущий перечень полок:", ", ".join(list(directories)))
delete_shelf()

def main():
    while True: 
        print()
        print("Введите команду: ")
        user_input = input()
        if user_input  == "p":
            doc_owner()
        elif user_input  == "s":
            doc_shelf()
        elif user_input == "l":
            all_doc()
        elif user_input == "as":
            new_shelf()
        elif user_input == "ds":
            delete_shelf()
        elif user_input == "q":
            print("Спасибо за работу!")
            break
```

```python
main()
```
