```python
documents = [
{"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
{"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
{"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
"1": ["2207 876234", "11-2"],
"2": ["10006"],
"3": []
}

def doc_owner():  
    print("", "Введите номер документа: ", sep="\n")
    user_input = input()
    print()
    for doc in documents:
        if user_input == doc["number"]:
            owner = doc["name"]
            break
    else: 
        owner = "Документ в базе не найден"
    print("Результат:", owner)
doc_owner()

def doc_shelf():  
    print("", "Введите номер документа: ", sep="\n")
    user_input = input()
    print()
    dir_list = list(directories)
    values_list = list(directories.values())
    i = 0
    for dir in directories:  
        if user_input in values_list[i]:
            shelf = dir_list[i]
            print("Результат:")
            print("Документ хранится на полке:", shelf)
            break
        else:
            i += 1
    else:
        print("Результат:", "Документ в базе не найден", sep="\n")
doc_shelf()

def all_doc():
    print("", "Результат:", sep="\n")
    dir_list = list(directories)
    values_list = list(directories.values())
    l = len(dir_list)
    for doc in documents:
        for i in range(0, l):
            if doc["number"] in values_list[i]:
                shelf = dir_list[i]
                print("№: {}, тип: {}, владелец: {}, полка хранения: {}".format(doc['number'], doc['type'], doc['name'], shelf))
all_doc()

def new_shelf():
    print("", "Введите номер полки: ", sep="\n")
    user_input = input()
    print()
    if user_input in list(directories):
        sorted_dir = sorted(int(dir) for dir in list(directories))
        print("Результат:")
        print("Такая полка уже существует. Текущий перечень полок: ", ", ".join(str(dir) for dir in sorted_dir))
    else:
        directories[user_input] = []
        sorted_dir = sorted(int(dir) for dir in list(directories))
        print("Полка добавлена. Текущий перечень полок: ", ", ".join(str(dir) for dir in sorted_dir))
new_shelf()

def delete_shelf():
    print()  
    print("Введите номер полки: ")
    user_input = input()
    print()
    if user_input in list(directories) and directories[user_input] == []:
        del(directories[user_input])
        sorted_dir = sorted(int(dir) for dir in list(directories))
        print("Результат: ") 
        print("Полка удалена. Текущий перечень полок:", ", ".join(str(dir) for dir in sorted_dir))
    elif user_input in list(directories) and directories[user_input] != []:
        sorted_dir = sorted(int(dir) for dir in list(directories))
        print("Результат: ") 
        print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:", ", ".join(str(dir) for dir in sorted_dir))
    else:
        sorted_dir = sorted(int(dir) for dir in list(directories))
        print("Результат: ") 
        print("Такой полки не существует. Текущий перечень полок:", ", ".join(str(dir) for dir in sorted_dir))
delete_shelf()

def new_doc():
    i = 0
    new_doc = {}
    new_shelf = []
    dir_list = list(directories)
    values_list = list(directories.values())
    new_dir_list = []
    
    print("", "Введите номер документа:", sep="\n")
    user_input1 = input()
    new_doc["number"] = user_input1
    new_dir_list.append(user_input1)
    
    print("", "Введите тип документа:", sep="\n")
    user_input2 = input()
    new_doc["type"] = user_input2
    
    print(" ", "Введите владельца документа:", sep="\n")
    user_input3 = input()
    new_doc["name"] = user_input3
    
    print("", "Введите полку для хранения:", sep="\n")
    user_input4 = input()
    l = len(dir_list)
    for i in range(0, l):
        if user_input4 in dir_list[i]:
            directories[dir_list[i]] = user_input1
        else: 
            directories[user_input4] = new_dir_list
    new_doc["shelf"] = user_input4
    
    documents.append(new_doc)
    
    print("", "Результат:", "", "Документ добавлен. Текущий список документов:", sep="\n")   
    dir_list = list(directories)
    l = len(dir_list)
    values_list = list(directories.values())
    for doc in documents:
        for i in range(0, l):
            if doc["number"] in values_list[i]:
                shelf = dir_list[i]
                print("№: {}, тип: {}, владелец: {}, полка хранения: {}".format(doc['number'], doc['type'], doc['name'], shelf))
new_doc()

def main():
    while True: 
        print("", "Введите команду: ", sep="\n")
        user_input = input()
        if user_input  == "p":
            doc_owner()
        elif user_input  == "s":
            doc_shelf()
        elif user_input == "l":
            all_doc()
        elif user_input == "as":
            new_shelf()
        elif user_input == "ds":
            delete_shelf()
        elif user_input == "ad":
            new_doc()
        elif user_input == "q":
            print("Спасибо за работу!")
            break
```

```python
main()
```
