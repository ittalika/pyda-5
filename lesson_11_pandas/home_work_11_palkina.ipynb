Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

data_movies = pd.read_csv('movies.csv')
data_ratings = pd.read_csv('ratings.csv')

data_joined = data_ratings.merge(data_movies, on='movieId', how='left')
data_joined[data_joined.rating == 5]['title'].value_counts().head(1)
```

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data_power = pd.read_csv('power.csv')

data_baltics = data_power[ (data_power['country'] == 'Latvia') | (data_power['country'] == 'Lithuania') | (data_power['country'] == 'Estonia') ]

data_baltics_by_year = data_baltics[ (data_baltics.year >= 2005) & (data_baltics.year <= 2010) ]
data_baltics_filtered = data_baltics_by_year[ (data_baltics_by_year.category == 4) | (data_baltics_by_year.category == 12) | (data_baltics_by_year.category == 21) ]
 
print("Суммарное потребление стран Прибалтики:", int(data_baltics_filtered[ data_baltics_filtered.quantity > 0 ]['quantity'].sum()))
```
